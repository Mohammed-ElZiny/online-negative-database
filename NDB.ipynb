```python
class NegativeDatabase:
    def __init__(self):
        self.NDB = set()  # Initialize an empty negative database

    def Empty_NDB_create(self):
        """Create an empty negative database.
        Pick dlog(l)e bit positions at random
2. for every possible assignment Vp of this positions{
3.
select k1 randomly 1 <= k1 <= l
4.
for j=1 to k1 {
5.
select an additional n distinct positions
6.
for every possible assignment Vq of these positions{
7.
Pick k2 bits at random from Vp · Vq .
8.
Create a entry for N DB with the k2
chosen bits and fill the remaining l − k2
positions with the don’t care symbol.}}}

        """
        self.NDB = set()

    def string_to_binary(self, s):
        """Convert a string to its binary representation."""
        return ''.join(format(ord(c), '08b') for c in s)

    def binary_complement(self, binary_str):
        """Compute the bitwise complement of a binary string."""
        return ''.join('1' if bit == '0' else '0' for bit in binary_str)

    def negative_pattern_generate(self, x):
        """
        Generate the negative representation for pattern x.
        Convert x to binary and store its complement in NDB.
        Create a random permutation π
2. for all specified bits bi in π(x)
3.
Let x0 be the same as π(x) but with bi flipped
4.
if x0 is subsumed by some string in π(N DB)
5.
π(x) ← π(x) − ith bit (set value to 0 ∗0 )
6.
Keep track of the ith bit in a set indicator vector (SIV)
7. Randomly choose 0 ≤ t ≤ |SIV |
8. R ← t randomly selected bits from SIV
9. Create a pattern Vk using π(x) and the bits indicated by R.
10. return π 0 (Vk )a
a
π 0 is the inverse permutation of π.
        """
        x_binary = self.string_to_binary(x)
        neg_repr = self.binary_complement(x_binary)
        self.NDB.add(neg_repr)

    def insert(self, x):
        """Insert a new pattern x into the negative database.
  Randomly choose 1 ≤ j ≤ l
2. for k = 1 to j do
3.
Randomly select from x at most n distinct unspecified bit positions
4.
for every possible bit assignment Bp of the selected positions
5.
x0 ← x · Bp
6.
y ← Negative Pattern Generate(x0 , N DB)
7.
add y to N DB
        """
        x_binary = self.string_to_binary(x)
        self.NDB.add(x_binary)

    def delete(self, x):
        """Remove pattern x from the negative database.
        Let Dx be all the strings in N DB that match x
2. Remove Dx from N DB.
3. for all y ∈ Dx
4.
for each unspecified position qi of y
5.
Create a new string y 0 using the specified bits of y and the
complement of the bit specified at the ith position of x.
6.
Insert(y 0 , N DB)"""
        x_binary = self.string_to_binary(x)
        self.NDB.discard(x_binary)

    def update(self, x, y):
        """
        Update the negative representation for an existing pattern x.
        Replace x with y in the database.
        """
        self.delete(x)
        self.insert(y)

# Example usage:
if __name__ == "__main__":
    ndb = NegativeDatabase()

    # Initialize with some initial data (optional)
    initial_data = {"apple", "banana", "cherry"}
    for item in initial_data:
        ndb.insert(item)

    # Insert a new pattern
    ndb.insert("date")

    # Generate negative representation for a pattern
    x = "bana?a"
    neg_repr = ndb.negative_pattern_generate(x)
    print(f"Negative representation for '{x}': {neg_repr}")

    # Delete a pattern
    ndb.delete("cherry")

    # Update a pattern
    ndb.update("banana", "pineapple")

    # Print the current negative database
    print("Current Negative Database:")
    for item in ndb.NDB:
        print(item)
```

```python
import random
import string
import math

class NegativeDatabase:
    def __init__(self):
        self.NDB = set()  # Initialize an empty negative database

    def Empty_NDB_create(self):
        """Create an empty negative database."""
        d = 20  # Number of random bit positions to pick
        l = 32  # Length of each name (number of bits)
        n = 5   # Number of additional distinct positions to select

        # Generate random names and convert them to binary
        names = [''.join(random.choice(string.ascii_letters) for _ in range(l)) for _ in range(20)]

        for name in names:
            binary_name = ''.join(format(ord(c), '08b') for c in name)

            # Pick dlog(l)e bit positions at random
            positions = random.sample(range(l), int(math.log2(l)))

            for position in positions:
                # Generate a random assignment Vp for the selected positions
                Vp = ''.join(random.choice(["0", "1"]) for _ in range(l))

                # Select k1 randomly
                k1 = random.randint(1, l)

                for _ in range(k1):
                    # Select an additional n distinct positions
                    additional_positions = random.sample(range(l), n)

                    for additional_position in additional_positions:
                        # Generate a random assignment Vq for the additional positions
                        Vq = ''.join(random.choice(["0", "1"]) for _ in range(l))

                        # Pick k2 bits at random from Vp * Vq
                        k2 = random.randint(1, l)
                        selected_bits = random.sample(range(l), k2)
                        chosen_bits = ''.join(Vp[bit] for bit in selected_bits)

                        # Create an entry for NDB with the chosen bits and fill the remaining positions with don't care symbols
                        entry = chosen_bits + '*' * (l - k2)
                        self.NDB.add(entry)

        print("Empty negative database created successfully.")

    def string_to_binary(self, s):
        """Convert a string to its binary representation."""
        return ''.join(format(ord(c), '08b') for c in s)

    def binary_complement(self, binary_str):
        """Compute the bitwise complement of a binary string."""
        return ''.join('1' if bit == '0' else '0' for bit in binary_str)

    def negative_pattern_generate(self, x):
        """
        Generate the negative representation for pattern x.
        Convert x to binary and store its complement in NDB.
        """
        binary_x = self.string_to_binary(x)
        complement_x = self.binary_complement(binary_x)
        return complement_x

    def insert_record(self, pattern):
        """
        inserting a record by deleting its pattern from ndb "delete_record"
        """
        binary_pattern = self.string_to_binary(pattern)

        # Find entries to delete
        entries_to_remove = set()
        for entry in self.NDB:
            matched = True
            for i, bit in enumerate(entry):
                if bit != '*' and bit != binary_pattern[i]:
                    matched = False
                    break
            if matched:
                entries_to_remove.add(entry)

        # Remove Dx from NDB
        self.NDB -= entries_to_remove

        # Reinsert strings represented by Dx except x
        for entry in entries_to_remove:
            for i, bit in enumerate(entry):
                if bit == '*':
                    new_entry = entry[:i] + ('1' if binary_pattern[i] == '0' else '0') + entry[i+1:]
                    self.NDB.add(new_entry)

        # binary_pattern = self.string_to_binary(pattern)

        # #find entries to delete
        # entries_to_remove = set()
        # for entry in self.NDB:
        #     matched = True
        #     for i, bit in enumerate(entry):
        #         if bit != '*' and bit != binary_pattern[i]:
        #             matched = False
        #             break
        #     if matched:
        #         entries_to_remove.add(entry)
        # entries_to_remove = set()
        # for entry in self.NDB:
        #     matched = True
        #     for i, bit in enumerate(entry):
        #         if bit != '*' and bit != binary_pattern[i]:
        #            matched = False
        #            break
        #         if matched:
        #            entries_to_remove.add(entry)

        #remove entries from negative database
        # for entry in entries_to_remove:
        #     self.NDB.remove(entry)
        self.save_to_file("ndb_file")


    def save_to_file(self,filename):
        """save NDB to a file"""
        try:
          with open(filename,'w') as file:
            for entry in self.NDB:
                file.write(entry +"\n")
          print(f"NDB saved to {filename}")
        except IOError:
          print("Error: cannot save to file!")


    def load_from_file(self,filename):
        """load NDB from a file"""
        try:
          with open(filename, 'r') as file:
            for line in file:
              self.NDB.add(line.strip())
          print(f"NDB loaded successfully from {filename}")

        except IOError:
          print("Error:file not found!")


    def query(self, pattern):
        """Query the negative database for a pattern."""
        binary_pattern = self.string_to_binary(pattern)

        for entry in self.NDB:
            matched = True
            for i, bit in enumerate(entry):
                if bit != '*' and bit != binary_pattern[i]:
                    matched = False
                    break
            if matched:
                print(f"Pattern '{pattern}' matched with entry '{entry}' in the negative database.")
                return

        print(f"No match found for pattern '{pattern}' in the negative database.")

    def delete_record(self, pattern):
        """Add a negative pattern to the NDB."""
        binary_pattern = self.string_to_binary(pattern)

        # Use the prefix algorithm to enhance space usage
        prefix_added = False
        for entry in list(self.NDB):
            common_prefix = self._find_common_prefix(binary_pattern, entry)
            if common_prefix:
                remaining_bits = binary_pattern[len(common_prefix):]
                if remaining_bits:
                    new_entry = common_prefix + '*' * (len(entry) - len(common_prefix))
                    self.NDB.add(new_entry)
                else:
                    self.NDB.remove(entry)
                prefix_added = True
                break

        if not prefix_added:
            self.NDB.add(binary_pattern)
        print(f"positive pattern '{pattern}' deleted.")
        self.save_to_file("ndb_file")

    def _find_common_prefix(self, pattern1, pattern2):
        """Find the common prefix between two patterns."""
        common_prefix = ''
        for bit1, bit2 in zip(pattern1, pattern2):
            if bit1 == bit2:
                common_prefix += bit1
            else:
                break
        return common_prefix

    def update(self, old_pattern, new_pattern):
        """
        Update the negative representation for an existing pattern in the database.
        """
        # Find and remove the old representation
        self.delete_record(old_pattern)
        self.insert_record(new_pattern)
        print(f"Negative representation for pattern '{old_pattern}' updated to '{new_pattern}'.")


# Example usage:
if __name__ == "__main__":
    ndb = NegativeDatabase()
    ndb.Empty_NDB_create()
    print(len(ndb.NDB))

    # Query the negative database
    pattern_to_query = "banana"
    ndb.query(pattern_to_query)

    #remove pattern from NDB (inserting to DB)
    pattern_to_remove = "banana"
    ndb.insert_record(pattern_to_remove)
    ndb.query(pattern_to_remove)
    print(len(ndb.NDB))

    # check for update with another pattern
    pattern2 = "strewberry"
    ndb.query(pattern2)

    ndb.insert_record(pattern2)
    ndb.query(pattern2)
    ndb.query("apple")
    print(len(ndb.NDB))

    ndb.update(pattern_to_remove, "apple")
    ndb.query("apple")
    ndb.query(pattern_to_remove)
    print(len(ndb.NDB))

```
